```python
import numpy as np
from classes.PanoramaProcessor import PanoramaProcessor
import sys
import os
project_path = r"c:\Users\RaimondoReggio\OneDrive - Net Service S.p.A\Desktop\GitHub\hammon-WP4\Street_view\rectified_facades_extraction_from_panorama"
sys.path.append(project_path)
from s3_library.S3Client import S3Client


def main():
    """
    Main function that configures and starts the panorama processor.
    """
    # Configuration
    root = 'rectification/'
    country_city = 'results/'
    plot_redundant = True
    save_directly = False
    new_count = 5
    
    # New parameters for the GeoDataFrame
    images_base_path = '2025/'

    s3_client = S3Client()

    for zone in s3_client.list_all_sub_folders('data', 'geodataframe/results'):
        file_names = s3_client.list_files('data', 'geodataframe/results/' + zone)
        geodataframe_path = [name for name in file_names if name.endswith('.geojson')][0]

        # Initialize and start the processor
        processor = PanoramaProcessor(
            root=root,
            country_city=country_city,
            zone=zone,
            plot_redundant=plot_redundant,
            save_directly=save_directly,
            new_count=new_count,
            geodataframe_path=geodataframe_path,
            images_base_path=images_base_path,
            s3_client=s3_client
        )
        
        processor.process_all_panoramas()


if __name__ == "__main__":
    # Uncomment to use a fixed seed for reproducibility
    # np.random.seed(1)
    main()
```
